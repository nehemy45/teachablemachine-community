# CREATE DATAFRAME

```python
import pandas as pd
import numpy as np
import time

dates = pd.date_range('today', periods=6)
num_arr = np.random.randn(6,4)
columns = ['A','B','C','D']
df1 = pd.DataFrame(num_arr, index=dates, columns=columns) # create dataframe
df1
```

```python
# create dataframe with dictionary array
data = {'animal':['cat','dog','chicken','snake','dog','cat'],'age':[12,23,np.nan,33,43,52],'vistis':[1,2,3,4,1,2],'priority':['yes','no','yes','no','no','yes']}
labels = ['a','b','c','e','f','g']
df2 = pd.DataFrame(data, index=labels)
df2
```
